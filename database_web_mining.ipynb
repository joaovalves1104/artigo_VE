```python
import requests
from bs4 import BeautifulSoup
import os

def get_dataset_pages(base_url):
    response = requests.get(base_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    links = soup.find_all('a', href=True)
    dataset_links = [link['href'] for link in links if '/dataset/' in link['href'] and link['href'].count('/') == 2]
    dataset_links = ['https://dadosabertos.ccee.org.br' + link for link in dataset_links if not link.startswith('http')]
    return list(set(dataset_links))  # Remove duplicatas

def get_csv_links(dataset_page_url):
    response = requests.get(dataset_page_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    links = soup.find_all('a', text='CSV')
    csv_links = [link['href'] for link in links if link['href'].endswith('.csv')]
    return csv_links

def download_csv(url, folder):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            file_name = url.split('/')[-1]
            file_path = os.path.join(folder, file_name)
            with open(file_path, 'wb') as file:
                file.write(response.content)
            print(f"Arquivo salvo: {file_path}")
        else:
            print(f"Falha no download: {url} com status {response.status_code}")
    except Exception as e:
        print(f"Erro ao baixar {url}: {e}")

script_dir = os.getcwd()
download_folder = os.path.join(script_dir, "ccee_csv_files")
if not os.path.exists(download_folder):
    os.makedirs(download_folder)

base_url = "https://dadosabertos.ccee.org.br/dataset/"
dataset_pages = get_dataset_pages(base_url)
print(f"Encontradas {len(dataset_pages)} páginas de datasets.")

for dataset_page in dataset_pages:
    print(f"Processando página de dataset: {dataset_page}")
    csv_links = get_csv_links(dataset_page)
    if csv_links:
        print(f"Encontrados {len(csv_links)} arquivos CSV na página {dataset_page}.")
        for link in csv_links:
            download_csv(link, download_folder)
    else:
        print(f"Nenhum arquivo CSV encontrado na página {dataset_page}.")
```

```python
import requests
from bs4 import BeautifulSoup
import os

def get_dataset_pages(base_url):
    response = requests.get(base_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    links = soup.find_all('a', href=True)
    dataset_links = [link['href'] for link in links if '/dataset/' in link['href'] and link['href'].count('/') == 2]
    dataset_links = ['https://dadosabertos.ccee.org.br' + link for link in dataset_links if not link.startswith('http')]
    return list(set(dataset_links))  # Remove duplicatas

def get_csv_links(dataset_page_url):
    response = requests.get(dataset_page_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    csv_links = []
    resources = soup.find_all('li', {'class': 'resource-item'})
    for resource in resources:
        a_tags = resource.find_all('a')
        for a in a_tags:
            if a.text.strip().lower() == 'csv':
                link = a['href']
                if not link.startswith('http'):
                    link = 'https://dadosabertos.ccee.org.br' + link
                csv_links.append(link)
                break
    return csv_links

def download_csv(url, folder):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            file_name = url.split('/')[-1]
            file_path = os.path.join(folder, file_name)
            with open(file_path, 'wb') as file:
                file.write(response.content)
            print(f"Arquivo salvo: {file_path}")
        else:
            print(f"Falha no download: {url} com status {response.status_code}")
    except Exception as e:
        print(f"Erro ao baixar {url}: {e}")

script_dir = os.getcwd()
download_folder = os.path.join(script_dir, "ccee_csv_files")
if not os.path.exists(download_folder):
    os.makedirs(download_folder)

base_url = "https://dadosabertos.ccee.org.br/dataset/"
dataset_pages = get_dataset_pages(base_url)
print(f"Encontradas {len(dataset_pages)} páginas de datasets.")

for dataset_page in dataset_pages:
    print(f"Processando página de dataset: {dataset_page}")
    csv_links = get_csv_links(dataset_page)
    if csv_links:
        print(f"Encontrados {len(csv_links)} arquivos CSV na página {dataset_page}.")
        for link in csv_links:
            download_csv(link, download_folder)
    else:
        print(f"Nenhum arquivo CSV encontrado na página {dataset_page}.")
```

```python
import csv
import requests
import os

def get_records_via_api(base_url, resource_id, limit=5, offset=0):
    search_url = f"{base_url}?resource_id={resource_id}&limit={limit}&offset={offset}"
    try:
        response = requests.get(search_url)
        response.raise_for_status()  # Vai levantar uma exceção para respostas 4xx/5xx
        data = response.json()
        return data['result']['records'], data['result']['total']
    except requests.RequestException as e:
        print(f"Erro ao acessar a API: {e}")
        return [], 0

def save_records_to_csv(records, download_folder, filename='output.csv'):
    keys = records[0].keys() if records else []
    file_path = os.path.join(download_folder, filename)
    with open(file_path, 'w', newline='', encoding='utf-8') as output_file:
        dict_writer = csv.DictWriter(output_file, keys)
        dict_writer.writeheader()
        dict_writer.writerows(records)
    print(f"Arquivo CSV salvo em {file_path}")

# Configuração de diretórios
script_dir = os.getcwd()
download_folder = os.path.join(script_dir, "ccee_csv_files")
os.makedirs(download_folder, exist_ok=True)

# Configuração da API
base_url = "https://dadosabertos.ccee.org.br/api/3/action/datastore_search"
resource_id = "5c64e360-0252-4849-9dbb-8a61cb2af8f0"
limit = 100  # Ajuste conforme necessário
offset = 0
total_records = []

# Recuperando todos os registros
while True:
    records, total = get_records_via_api(base_url, resource_id, limit, offset)
    if not records:
        break
    total_records.extend(records)
    offset += len(records)
    if offset >= total:
        break

# Salvando todos os registros em CSV
if total_records:
    save_records_to_csv(total_records, download_folder)
else:
    print("Nenhum registro encontrado para baixar.")
```

```python
import requests
import csv
import os

def get_records_via_api(base_url, resource_id, offset=0, limit=100):
    """Busca registros na API CKAN com suporte a paginação."""
    records = []
    while True:
        search_url = f"{base_url}?resource_id={resource_id}&offset={offset}&limit={limit}"
        response = requests.get(search_url)
        data = response.json()
        batch_records = data['result']['records']
        if not batch_records:
            break
        records.extend(batch_records)
        offset += limit
    return records

def save_records_to_csv(records, download_folder):
    """Salva registros em arquivos CSV separados."""
    for record in records:
        filename = f"{record['id']}.csv"  # Assumindo que 'id' é um identificador único
        file_path = os.path.join(download_folder, filename)
        keys = record.keys()
        with open(file_path, 'w', newline='', encoding='utf-8') as file:
            dict_writer = csv.DictWriter(file, keys)
            dict_writer.writeheader()
            dict_writer.writerow(record)
        print(f"Arquivo {filename} salvo com sucesso em {file_path}")

# Configuração inicial
script_dir = os.getcwd()
download_folder = os.path.join(script_dir, "ccee_csv_files")
if not os.path.exists(download_folder):
    os.makedirs(download_folder)

# Configuração da API
base_url = "https://dadosabertos.ccee.org.br/api/3/action/datastore_search"
resource_id = "5c64e360-0252-4849-9dbb-8a61cb2af8f0"  # Exemplo de ID de recurso

# Processamento
records = get_records_via_api(base_url, resource_id)
if records:
    save_records_to_csv(records, download_folder)
else:
    print("Nenhum registro encontrado para baixar.")
```
